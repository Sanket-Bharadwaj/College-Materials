# 1. Write a Python function basic salary that accepts two parameters: hourly rate and hours worked per week. The function should calculate the basic salary per month (assuming a month has 4 weeks). If the hours worked per week exceed 40, create another function overtime salary, where every extra hour worked is paid at 1.5 times the normal hourly rate. Finally, create another function total salary that returns the sum of the basic salary and overtime.

```python
def basic_salary(hourly_rate, hours_per_week):
    if hours_per_week > 40:
        normal_hours = 40
    else:
        normal_hours = hours_per_week
    return hourly_rate * normal_hours * 4

def overtime_salary(hourly_rate, hours_per_week):
    if hours_per_week > 40:
        extra_hours = hours_per_week - 40
        return extra_hours * hourly_rate * 1.5 * 4
    else:
        return 0

def total_salary(hourly_rate, hours_per_week):
    return basic_salary(hourly_rate, hours_per_week) + overtime_salary(hourly_rate, hours_per_week)

rate = float(input("Enter hourly rate: "))
hours = float(input("Enter hours worked per week: "))
print("Total salary per month:", total_salary(rate, hours))
```

# 2. Create a function tax amount that shows how much taxes are deducted from the basic salary. Taxes are applied as follows:


*   If the salary is less than Rs. 60,000/-, deduct 10% as tax.
*   If the salary is between Rs. 60,000/- and Rs. 85,000/-, deduct 15% as tax.
*   If the salary is more than Rs. 85,000/-, deduct 20% as tax.

```python
def basic_salary(hourly_rate, hours_per_week):
    if hours_per_week > 40:
        normal_hours = 40
    else:
        normal_hours = hours_per_week
    return hourly_rate * normal_hours * 4

def overtime_salary(hourly_rate, hours_per_week):
    if hours_per_week > 40:
        extra_hours = hours_per_week - 40
        return extra_hours * hourly_rate * 1.5 * 4
    else:
        return 0

def total_salary(hourly_rate, hours_per_week):
    return basic_salary(hourly_rate, hours_per_week) + overtime_salary(hourly_rate, hours_per_week)

def tax_amount(basic):
    if basic < 60000:
        return basic * 0.10
    elif basic <= 85000:
        return basic * 0.15
    else:
        return basic * 0.20

rate = float(input("Enter hourly rate: "))
hours = float(input("Enter hours worked per week: "))
basic = basic_salary(rate, hours)
print("Basic salary:", basic)
print("Tax amount:", tax_amount(basic))
print("Total salary (including overtime):", total_salary(rate, hours))
```

# 3. Using the function basic salary from Question 1, write another function gross salary that calculates the gross salary of an employee. This function should accept basic salary as input (output from Question 1), consider a fixed value of allowances (e.g., 20% of basic salary), and return the gross salary (basic salary + allowances - tax).

```python
def basic_salary(hourly_rate, hours_per_week):
    if hours_per_week > 40:
        normal_hours = 40
    else:
        normal_hours = hours_per_week
    return hourly_rate * normal_hours * 4

def overtime_salary(hourly_rate, hours_per_week):
    if hours_per_week > 40:
        extra_hours = hours_per_week - 40
        return extra_hours * hourly_rate * 1.5 * 4
    else:
        return 0

def total_salary(hourly_rate, hours_per_week):
    return basic_salary(hourly_rate, hours_per_week) + overtime_salary(hourly_rate, hours_per_week)

def tax_amount(basic):
    if basic < 60000:
        return basic * 0.10
    elif basic <= 85000:
        return basic * 0.15
    else:
        return basic * 0.20

def gross_salary(basic):
    allowance = basic * 0.20
    tax = tax_amount(basic)
    return basic + allowance - tax

rate = float(input("Enter hourly rate: "))
hours = float(input("Enter hours worked per week: "))
basic = basic_salary(rate, hours)
print("Basic salary:", basic)
print("Gross salary:", gross_salary(basic))
```

#4. Using the gross salary function from Question 3, write a function salary bracket that categorizes the employee’s gross salary into one of the following brackets:

* ”Middle income” if gross salary is between Rs. 50,000/- and Rs. 80,000/-.

```python
def basic_salary(hourly_rate, hours_per_week):
    if hours_per_week > 40:
        normal_hours = 40
    else:
        normal_hours = hours_per_week
    return hourly_rate * normal_hours * 4

def overtime_salary(hourly_rate, hours_per_week):
    if hours_per_week > 40:
        extra_hours = hours_per_week - 40
        return extra_hours * hourly_rate * 1.5 * 4
    else:
        return 0

def total_salary(hourly_rate, hours_per_week):
    return basic_salary(hourly_rate, hours_per_week) + overtime_salary(hourly_rate, hours_per_week)

def tax_amount(basic):
    if basic < 60000:
        return basic * 0.10
    elif basic <= 85000:
        return basic * 0.15
    else:
        return basic * 0.20

def gross_salary(basic):
    allowance = basic * 0.20
    tax = tax_amount(basic)
    return basic + allowance - tax

def salary_bracket(gross):
    if gross < 50000:
        return "Low income"
    elif gross <= 80000:
        return "Middle income"
    else:
        return "High income"

rate = float(input("Enter hourly rate: "))
hours = float(input("Enter hours worked per week: "))
basic = basic_salary(rate, hours)
gross = gross_salary(basic)
print("Basic salary:", basic)
print("Gross salary:", gross)
print("Salary bracket:", salary_bracket(gross))
```

# 5. Take three different sets of employee names, hourly rates and hours worked per week as user input. Write a Python function employee report that generates a formatted report of all employees’ salary details. This function should print the employee names, basic salaries, gross salaries, tax amounts, and salary brackets in a readable format.

```python
def basic_salary(base_rate, hours_per_week):
    if hours_per_week > 40:
        normal_hours = 40
    else:
        normal_hours = hours_per_week
    return base_rate * normal_hours * 4

def overtime_salary(base_rate, hours_per_week):
    if hours_per_week > 40:
        extra_hours = hours_per_week - 40
        overtime_rate = base_rate * 1.5
        return overtime_rate * extra_hours * 4
    else:
        return 0

def total_salary(base_rate, hours_per_week):
    return basic_salary(base_rate, hours_per_week) + overtime_salary(base_rate, hours_per_week)

def tax_amount(basic):
    if basic < 60000:
        return basic * 0.10
    elif basic <= 85000:
        return basic * 0.15
    else:
        return basic * 0.20

def gross_salary(basic):
    allowance = basic * 0.20
    tax = tax_amount(basic)
    return basic + allowance - tax

def salary_bracket(gross):
    if gross < 50000:
        return "Low income"
    elif gross <= 80000:
        return "Middle income"
    else:
        return "High income"

def employee_report():
    employees = []
    for i in range(3):
        name = input("Enter employee name: ")
        rate = float(input("Enter base hourly rate: "))
        hours = float(input("Enter hours worked per week: "))
        basic = basic_salary(rate, hours)
        total = total_salary(rate, hours)
        gross = gross_salary(total)
        tax = tax_amount(total)
        bracket = salary_bracket(gross)
        employees.append((name, basic, total, gross, tax, bracket))
        print()

    print("Employee Salary Report\n")
    print(f"{'Name':<12}{'Basic Salary':<15}{'Total Salary':<15}{'Gross Salary':<15}{'Tax':<10}{'Bracket'}")
    print()
    for emp in employees:
        print(f"{emp[0]:<12}{emp[1]:<15.2f}{emp[2]:<15.2f}{emp[3]:<15.2f}{emp[4]:<10.2f}{emp[5]}")

employee_report()
```
