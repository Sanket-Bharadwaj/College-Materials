9. format_2d_array function.

```python
def format_2d_array(arr):
    w=max(len(str(x)) for r in arr for x in r)
    return '\n'.join(' '.join(f"{x:>{w}}" for x in r) for r in arr)
print(format_2d_array([[1,22],[333,4]]))
```

```python
import numpy as np
def median(a):return np.median(a)
def mode(a):u,c=np.unique(a,return_counts=True);return u[c.argmax()]
print(median([1,2,3]))
print(mode([1,1,2,3]))
```

21. Series vowel filtering.

```python
import pandas as pd
L=['Cry','Apple','Orange','Sky','Banana']
s=pd.Series(L)
v=s[s.str.contains('[AEIOUaeiou]')]
st=s[s.str.contains('^[AEIOUaeiou]')]
print(v)
print(st)
```

23. DataFrame tasks.

```python
import pandas as pd
df=pd.DataFrame({'Maxine':[1,2,3],'James':[4,5,6],'Amanda':[7,8,9]})
df2=pd.DataFrame({'Maxine':[1,2,3],'James':[4,5,6],'Amanda':[7,8,9]},index=['Morning','Afternoon','Evening'])
print(df2['Maxine'])
print(df2.loc['Morning'])
print(df2.loc[['Morning','Evening']])
print(df2[['Amanda','Maxine']])
print(df2.loc[['Morning','Afternoon'],['Amanda','Maxine']])
print(df2.describe())
print(df2.T)
print(df2.sort_index(axis=1))
```
