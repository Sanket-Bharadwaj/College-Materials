### 18. analyze_sets

```
def analyze_sets(l1,l2):
    s1,s2=set(l1),set(l2)
    sd=s1^s2
    res=[]
    for x in sd: res.append(x*2 if x%2==0 else x+5)
    return sorted(res)
print(analyze_sets([1,2,3],[3,4,5]))
```

### 19. unique_pairs

```
def unique_pairs(words):
    words=[w.lower() for w in words if len(w)>=4]
    res=set()
    for i in range(len(words)):
        for j in range(i+1,len(words)):
            if set(words[i]).isdisjoint(words[j]):res.add(tuple(sorted((words[i],words[j]))))
    return res
print(unique_pairs(['apple','dogs','bird','fish','zebra','lion']))
```
