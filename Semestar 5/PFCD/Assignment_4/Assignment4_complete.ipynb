## Q1. Create a list of N random numbers and find sum and average.

```
import random
n = 5
a = [random.randint(1,10) for i in range(n)]
print(sum(a))
print(sum(a)/n)
```

## Q2. Rotate three values using a function.

```
def rotate(a,b,c):
    return (c,a,b)
a,b,c = "Doug",22,1984
a,b,c = rotate(a,b,c)
print(a,b,c)
```

## Q7. Create cumulative sum list.

```
a = [1,2,3,4]
c = []
s = 0
for i in a:
    s += i
    c.append(s)
print(c)
```

## Q8. Create list of lists with multiples.

```
n = 3
a = []
for i in range(1,n+1):
    a.append([i*j for j in range(1,6)])
print(a)
```

## Q10. Sum of elements in tuple of tuples.

```
t = ((1,2),(3,4),(5,6))
s = 0
for i in t:
    for j in i:
        s += j
print(s)
```

## Q19. Demonstrate shallow and deep copy.

```
import copy
a = [['Shallow',2,3],[4,5,6]]
b = copy.copy(a)
c = copy.deepcopy(a)
a[0][0] = 'Deep'
print(b)
print(c)
```

## Q20. Find statistics of ratings.

```
import statistics
a = [1,2,5,4,3,5,2,1,3,3,1,4,3,3,3,2,3,3,2,5]
print(min(a), max(a), max(a)-min(a))
print(statistics.mean(a))
print(statistics.median(a))
print(statistics.mode(a))
```
