Replace repeats with *.

```python
s='balloon'
r=s[0]
for i in range(1,len(s)): r+=('*' if s[i]==s[i-1] else s[i])
print(r)
```

Next vowel.

```python
s='education';v='aeiou'
print(''.join(v[(v.index(c)+1)%5] if c in v else c for c in s))
```

Rotational palindrome.

```python
s='aab'
print(any((s[i:]+s[:i])==(s[i:]+s[:i])[::-1] for i in range(len(s))))
```

Vowels and consonants.

```python
s='abcde';v='aeiou'
print(sum(c in v for c in s), sum(c.isalpha() and c not in v for c in s))
```

```python
s='bad bed bid'
print([w for w in s.split() if w[0]=='b' and w[-1]=='d'])
```

3-letter combos.

```python
import itertools
print([''.join(p) for p in set(itertools.permutations('abcde',3))])
```
